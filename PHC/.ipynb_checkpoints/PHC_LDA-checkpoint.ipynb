#### Data Preprocessing

```python
# seperate the data and label

# data = df.drop(['Color', 'blood_hemoglobin'], axis=1)
df = df.dropna()

data = df.drop(['Color'], axis=1)



label = df.Color
```

#### Label Binarization

#### Training and Test Data

```python
# seperating traing and test data
data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=0.2, random_state=2)

print(data.shape, data_train.shape, data_test.shape)
```

```python
# data scaling
sc = StandardScaler()
data_train = sc.fit_transform(data_train)
data_test = sc.transform(data_test)
```

### Performing LDA

```python
lda = LDA(n_components=3)
data_train = lda.fit_transform(data_train, label_train)
data_test = lda.transform(data_test)
```

### determining n_component

```python
lda_var_ratios = lda.explained_variance_ratio_
```

```python
def select_n_components(var_ratio, goal_var: float) -> int:
    # Set initial variance explained so far
    total_variance = 0.0
    
    # Set initial number of features
    n_components = 0
    
    # For the explained variance of each feature:
    for explained_variance in var_ratio:
        
        # Add the explained variance to the total
        total_variance += explained_variance
        
        # Add one to the number of components
        n_components += 1
        
        # If we reach our goal level of explained variance
        if total_variance >= goal_var:
            # End the loop
            break
            
    # Return the number of components
    return n_components
```

```python
select_n_components(lda_var_ratios, 0.98)
```

### Model Training
#### Random Forest  Classifier

```python
model = RandomForestClassifier(max_depth = 15, random_state=0)

model.fit(data_train, label_train)
```

### Model Evaluation
#### Accuracy Score

```python
# accuracy test on data
data_test_prediction = model.predict(data_test)
test_data_accuracy = accuracy_score(data_test_prediction, label_test)

print('Accuracy : ', test_data_accuracy* 100)
```

```python
### testing accuracy with confusion metrics

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(label_test, data_test_prediction)
print(cm)
print('Accuracy' + str(accuracy_score(label_test, data_test_prediction)))
```
