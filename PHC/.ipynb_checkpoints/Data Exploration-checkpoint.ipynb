```python
#calculating correlation
correlation = df.corr()

# constructing a heatmap to undestand the correlation between the columns
plt.figure(figsize=(15,15))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':12}, cmap = 'OrRd')
```

```python
#Correlation with output variable
cor_target = abs(correlation["Color"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0]
relevant_features.sort_values(ascending=False)
```

```python
import seaborn as sns
data_values = df['BloodGlucose']
sns.boxplot(x=data_values)


from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df))
print(z)
threshold = 3
print(np.where(z > threshold))
```

```python
out = pd.cut(df['Age '], bins=[0, 18, 44, 64, 84, 120], labels=['0-18', '18-44', '44-64', '64-84', '84+'],include_lowest=False)
out_norm = out.value_counts(sort=False, normalize=True).mul(100)
ax = out_norm.plot.bar(rot=0, color="b", figsize=(8,6))
ax.bar_label(ax.containers[0])
plt.ylabel("percentage")
plt.xlabel("age group")

plt.show()
```
