```python
import os
import numpy as np
import pandas as pd
import warnings
import random
from math import sqrt
from collections import Counter
```

```python
def k_nearest_naighbor(data, predict, k=3):
    if len(data) >= k:
        warning.warn('K is set to a value less than total voting group! IDIOT**!!  ')
    distances =[]
    for group in data:
        for features in data[group]:
            euclidean_distance = np.linalg.norm(np.array(features) - np.array(predict))
            distances.append([euclidean_distance, group])
    
    votes = [i[1] for i in sorted(distances)[:k]]
    vote_result = Counter(votes).most_common(1)[0][0]  
    confidence = Counter(votes).most_common(1)[0][1] / k   
    
    return vote_result , confidence
```

```python

df = pd.read_csv('../datasets/breast-cancer-wisconsin.data')
```

```python
# replacing missing parameters with a null
df.replace('?', np.nan , inplace=True)
print(df.isna().sum())

# replacing missing null with a value -9999
df.replace(np.nan, -99999, inplace=True)
df.isin(['-99999']).sum(axis=0)


# #removing useless columns
df = df.drop('id',axis=1)
df.head()
```

```python
full_data = df.astype(float).values.tolist()
random.shuffle(full_data)
```

```python
test_size = 0.4
train_set = {2:[], 4:[]}
test_set = {2:[], 4:[]}
train_data = full_data[:-int(test_size*len(full_data))]
test_data = full_data[-int(test_size*len(full_data)) :]
```

```python
for i in train_data:
    train_set[i[-1]].append(i[:-1])
    print(train_set[i[-1]])
```

```python
for i in test_data:
    test_set[i[-1]].append(i[:-1])
```

```python
correct = 0
total = 0
for group in test_set:
    for data in test_set[group]:
#         print(data)
        vote , confidence = k_nearest_naighbor(train_set, data, k=5)
        if group == vote:
            correct += 1
        else:
            print(confidence)
        total += 1
print(correct/total)
print(correct)
print(total)
```
