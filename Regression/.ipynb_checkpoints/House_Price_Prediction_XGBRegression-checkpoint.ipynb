# Boston House Price Prediction 

### Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

### Importing Boston Housing Dataset

```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset for the Pandas DataFrame
df = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
df.head()
```

```python
# add the target column to the DataFrame
df['PRICE'] = house_price_dataset.target
```

```python
# checking the number of rows and columns
print(f"Rows = {df.shape[0]}")
print(f"Columns = {df.shape[1]}")
```

```python
# check for missing values in each column
df.isnull().sum()
```

```python
# statistical measures of the dataset
df.describe()
```

### Understanding the correlation between various features in the dataset


1. Positive Correlation
2. Negative Correlation

```python
correlation = df.corr()
```

```python
# constructiong a heatmap to understand the correlation between columns
plt.figure(figsize = (15, 15))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':12}, cmap='OrRd')
```

### Splitting the data and Target

```python
X = df.drop('PRICE', axis=1)
Y = df['PRICE']
```

### Splitting the data into Training Data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
# checking the number of rows and columns for Trainig and Testing DataSets
print(f"Full Dataset (Rows={X.shape[0]}, columns={X.shape[1]})")
print(f"Training (Rows={X_train.shape[0]},columns={X_train.shape[1]})")
print(f"Testing (Rows={X_test.shape[0]}, columns={X_test.shape[1]})")
```
