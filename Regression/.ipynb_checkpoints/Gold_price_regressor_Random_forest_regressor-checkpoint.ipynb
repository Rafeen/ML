# Gold Price Prediction

### Import Dependencies

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

### Data Collection and Processing

```python
# Loading the csv data to a Pandas Dataframe
dirname = os.path.dirname("__file__")
filename = os.path.join(dirname, 'Datasets/gld_price_data.csv')
df = pd.read_csv(filename)
df.head()
```

```python
# checking row and columns
print(df.shape)
```

```python
# checking number of missing values
df.isnull().sum()
```

```python
# getting statistical measures of the data
df.describe()
```

#### Correlation
    1.Positive Correlation
    2.Negative Correlation

```python
correlation = df.corr()
```

```python
# constructiong a heatmap to understand the correlation between columns
plt.figure(figsize = (12, 12))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':12}, cmap='OrRd')
```

```python
# correlation values of GLD
print(correlation.GLD)
```

```python
# check the distribution of the GLD price

sns.distplot(df['GLD'], color="orange")
```

#### Seperating Test and training data

```python
# Splitting the features and the Label
data = df.drop(['GLD', 'Date'], axis=1)
label = df.GLD
```

```python
# seperating training and testing data
data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=0.2, random_state=2)
```

```python
# getting the shape of train and test data
print(data.shape, data_train.shape, data_test.shape)
```

### Model Training
#### Random Forest Regressor

```python
model = RandomForestRegressor()
```

```python
# training the model with training data
model.fit(data_train, label_train)
```

### Evaluation
#### Prediction on training data

```python
train_data_prediction = model.predict(data_train)
```

```python
# R squared Error
r2_score = metrics.r2_score(label_train, train_data_prediction)
print("R squared error value: ",r2_score)

mea_score = metrics.mean_absolute_error(label_train, train_data_prediction)
print("Mean error average:", mea_score)
```

#### Compare the actual value and the predicted value for training data

```python
label_test = list(label_test)
```

```python
plt.figure(figsize = (12, 12))
plt.plot(label_test, color='blue', label= "Actual Value")
plt.plot(train_data_prediction, color='red', label= "Predicted Value")
plt.title('Actual price Vs Predicted Price')
plt.xlabel('Number of Values')
plt.xlabel('GLD Price')
plt.legend()
plt.show()
```

#### Prediction on testing data

```python
test_data_prediction = model.predict(data_test)
```

```python
# R squared Error
r2_score = metrics.r2_score(label_test, test_data_prediction)
print("R squared error value: ",r2_score)

mea_score = metrics.mean_absolute_error(label_test, test_data_prediction)
print("Mean error average:", mea_score)
```

```python
plt.scatter(label_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
max_value = max([*label_test, *test_data_prediction])
min_value = min([*label_test, *test_data_prediction])
plt.plot([min_value,max_value],[min_value,max_value], 'red', linewidth=2)

plt.title("Actual Prices Vs Predicted Prices")
plt.show()
```
